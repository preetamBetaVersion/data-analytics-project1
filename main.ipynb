Q3) Mention the total no of nations who participated in each olympics game?

```python
data = dataset1[['Games', 'NOC']]
data = data.sort_values('Games')
data1 = data.groupby(by='Games', as_index=False).agg({'NOC' : pd.Series.nunique})
data1.columns = ['Game', 'Total countries']
data1
```

<hr>

Q4) Which year saw the highest and lowest no of countries participating in olympics?

```python
print('Game with highest number of countries participating:')
print(data1.max().to_string())
print('\n')
print('Game with lowest number of countries participating:')
print(data1.min().to_string())
```

Q5) Which nation has participated in all of the olympic games?

Q10) Find the Ratio of male and female athletes participated in all olympic games.

```python
new_df = dataset1[['Sex']]
number_of_males_in_all_og = new_df.loc[new_df['Sex']=='M'].count()
number_of_females_in_all_og = new_df.loc[new_df['Sex']=='F'].count()
ratio = int(number_of_males_in_all_og)/int(number_of_females_in_all_og)
print('The ratio of male to female athletes in all olympic games is ' + str(ratio))
```

Q11) Fetch the top 5 athletes who have won the most gold medals.

```python
from numpy import size


all_medal_athletes = dataset1[['Name','Team','Medal']].dropna()
all_gold_athletes = all_medal_athletes.loc[all_medal_athletes['Medal']=='Gold']
# to convert a series into a dataframe, we used the to_frame() method
all_gold_athletes_names = all_gold_athletes['Name'].to_frame()
all_gold_athletes_names = all_gold_athletes_names.pivot_table(index=['Name'], aggfunc='size').sort_values(ascending=False)
all_gold_athletes_names.to_frame().head(6)
```

**Note** : There is one player with 23 medals, one with 10 and 4 with 9. So to decide, which are the top 5 will need more criteria on which basis the categorization of the best 3 amongst the 4 players who would make into the list of the top 5 athletes.

Q12) Fetch the top 5 athletes who have won the most medals (gold/silver/bronze).

Q13) Fetch the top 5 most successful countries in olympics. Success is defined by no of medals won.

```python
top_five_teams = dataset1[['Team', 'Medal']].dropna()
top_five_teams.groupby('Team').count().sort_values(by='Medal', ascending=False).head(5)
```

Q14) List down total gold, silver and bronze medals won by each country.

```python
teams_with_medals = dataset1[['Team','NOC','Medal']].dropna().sort_values(by='Team')
c = pd.crosstab(teams_with_medals['Team'], teams_with_medals['Medal'])
c.index.name = 'Name of country'
c.sort_values(by='Gold', ascending=False)
```
